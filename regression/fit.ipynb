```python
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt
from scipy import optimize
%matplotlib.inline
```

```python
# y = [392677, 386303, 383907]
y = [493530, 488346, 488031]
y = np.array(y)
# x = [1671070575, 1671093497, 1671101363]
x = [1671329776, 1671364817, 1671367263]
x = np.array(x)
```

```python
def ax_plus_b(x, a, b):
    return a * x + b
```

```python
param_a, param_b = optimize.curve_fit(ax_plus_b, x, y)[0]
param_a, param_b
```
